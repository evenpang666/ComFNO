```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
```

```python
c1 = 1
c2 = 1
c3 = 1
c4 = 1
c5 = 1
c6 = 1
eps = 1/2**5

x = np.linspace(0,1,101)
y = c1*np.exp(-x/eps)+c2*np.exp((x-1)/eps)+(c3*np.sin(-x/eps)+c4*np.cos(-x/eps))*np.exp(x) + (c5*np.sin((x-1)/eps)+c6*np.cos((x-1)/eps))*np.exp(x)
fig = plt.figure(figsize=(8,5), dpi=150)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('u')

plt.subplot(2,2,1)
plt.plot(x,c1*np.exp(-x/eps),label="$e^{-x/\epsilon}$")
plt.plot(x,c2*np.exp((x-1)/eps),label="$e^{(x-1)/\epsilon}$")
plt.xlabel('x')
plt.ylabel('u')
plt.legend()
plt.tight_layout()

plt.subplot(2,2,2)
plt.plot(x,c3*np.sin(-x/eps),label="$sin(-x/\epsilon)$")
plt.plot(x,c4*np.cos(-x/eps),label="$cos(-x/\epsilon)$")
plt.plot(x,c5*np.sin((x-1)/eps),label="$sin((x-1)/\epsilon)$")
plt.plot(x,c6*np.cos((x-1)/eps),label="$cos((x-1)/\epsilon)$")
plt.xlabel('x')
plt.ylabel('u')
plt.legend()
plt.tight_layout()
plt.show()

fig.savefig("img/test_1.png")
```

```python
x = np.linspace(0,1,101)
eps = 1/2**5
X, Y = np.meshgrid(x, x)
Z = np.exp((X-1)/eps)+np.exp((Y-1)/eps)-np.exp((X-1)/eps)*np.exp((Y-1)/eps)

fig = plt.figure(figsize=(8,3),dpi=150)
ax1 = fig.add_subplot(1, 2, 1)
im1 = ax1.pcolormesh(X, Y, Z, shading='auto')
ax1.set_xlabel("y")
ax1.set_ylabel("x")
fig.colorbar(im1)

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
surf = ax2.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow'))
ax2.set_xlabel("y")
ax2.set_ylabel("x")
ax2.set_zlabel("u")
fig.colorbar(surf, shrink=0.5, aspect=10,pad=0.2)
plt.tight_layout()
plt.show()

fig.savefig("img/test_2.png")
```
